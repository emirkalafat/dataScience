```python
sns.displot(data=df, x='age', kde=True) #datasetteki yaş dağılımını gösteren barplot
```

```python
sns.relplot(data=df, x='duration', y='age', kind='scatter')
#Yaş ile kredi süresi arasındaki bağlantıyı inceleyen scatterplot
```

```python
sns.relplot(data=df, x='duration', y='age', kind='scatter', hue='job')
#aynı gösterimin iş dallarıyla olan dağılımı
#iş dalları fazla olduğu için anlaşılması zor bir gösterim oldu.
```

```python
sns.relplot(data=df, x='duration', y='age', kind='scatter', hue='job',col='education')
#aynı gösterimin iş dallarıyla olan dağılımı
#yukarıdaki tabloya aynı zamanda education(eğitim) feature'ı ile ilgli gösterim de ekledim
```

```python
sns.displot(data=df,
            x='age',hue='education')
#yaş dağılımını eğitim durumuna göre gösteren barplot
```

```python
sns.displot(data=df, x="balance", y="duration", kind="kde")
```

```python
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True)
```

```python
sns.violinplot(data=df, x="education", y="age",)
```

```python
sns.violinplot(data=df, x="education", y="balance",)
```

```python
sns.relplot(data=df, x='campaign', y='age', kind='scatter')
```

```python
sns.displot(data=df, x="age", y="balance",hue='contact',bins=200,)
```
